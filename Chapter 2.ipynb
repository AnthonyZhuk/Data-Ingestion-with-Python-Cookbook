```python
import os
import gnupg


gpg = gnupg.GPG(gnupghome='gpghome')
input_data = gpg.gen_key_input(
    name_email='testgpguser@mydomain.com',
    passphrase='mypassphrase')
key = gpg.gen_key(input_data)
```

```python
print(key)
```

```python
str(key)
```

```python
ascii_armored_public_keys = gpg.export_keys(str(key))
ascii_armored_private_keys = gpg.export_keys(str(key), True, passphrase="mypassphrase",)
with open('mykeyfile.asc', 'w') as f:
    f.write(ascii_armored_public_keys)
    f.write(ascii_armored_private_keys)
```

```python
#importing 

gpg = gnupg.GPG(gnupghome='gpghome')
key_data = open('mykeyfile.asc').read()
import_result = gpg.import_keys(key_data)
print(import_result.results)
```

```python
# encrypt

with open('spotify_data.csv', 'rb') as f:
    status = gpg.encrypt_file(
        f, recipients=['testgpguser@mydomain.com'],
        output='spotify_data.csv.gpg')

print('ok: ', status.ok)
print('status: ', status.status)
print('stderr: ', status.stderr)
```

```python
with open('spotify_data.csv.gpg', 'rb') as f:
    status = gpg.decrypt_file(f, passphrase='mypassphrase', output='spotify_data.csv')

print('ok: ', status.ok)
print('status: ', status.status)
print('stderr: ', status.stderr)
```
